```python
import pandas as pd
import os
from tqdm.notebook import tqdm
```

# Autism

```python
asd_data = pd.read_csv("data/gwas_statistics/autism/PGC.ASD.euro.all.25Mar2015.txt", sep="\t")
asd_data
```

```python
asd_data.sort_values(by="P")
```

```python
with open("data/gwas_statistics/autism/pegasus_autism_input_data.txt", "w") as f:
    for i, row in tqdm(asd_data.iterrows()):
        if not row["SNP"].startswith("rs"):
            continue
        f.write("{}\t{}\n".format(row["SNP"], row["P"]))
```

# Merging results

```python
autism_results = [f for f in os.listdir("PEGASUS/") if "autism_chr" in f]
autism_results = sorted(autism_results, key=lambda x: int(x.split("chr")[1].split(".")[0]))
autism_results

res = pd.DataFrame()
for fname in autism_results:
#     if "autism_chr" in fname:
#         print(fname)
    df = pd.read_csv("PEGASUS/"+fname, sep=" ") 
    res = pd.concat((res, df))
    
res.to_csv("processed_data/gwas_gene_pvals/autism/PEGASUS_autism_gwas_data.csv", index=False)
res
```

```python
sch_results = [f for f in os.listdir("PEGASUS/") if "schizophrenia_chr" in f]
sch_results = sorted(sch_results, key=lambda x: int(x.split("chr")[1].split(".")[0]))
sch_results

res = pd.DataFrame()
for fname in sch_results:
#     if "autism_chr" in fname:
#         print(fname)
    df = pd.read_csv("PEGASUS/"+fname, sep=" ") 
    res = pd.concat((res, df))
    
res.to_csv("processed_data/gwas_gene_pvals/schizophrenia/PEGASUS_schizophrenia_gwas_data.csv", index=False)
res
```
